```python
#To load different relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To have a basic overview of the dataset
df = pd.read_csv('SIPRI_Milex (3).csv')
df.head(15)
```

```python
print(df.columns)
```

```python
##Percentage increase in world military expenditures from 1988 to 2024

# Calculate the percentage increase in military expenditures from 1988 to 2024
df['Percentage_Increase_1988_2024'] = ((df['2024'] - df['1988']) / df['1988']) * 100

# Select and display relevant columns
df_percentage_increase = df[['Region', '1988', '2024', 'Percentage_Increase_1988_2024']]
df_percentage_increase.sort_values(by='Percentage_Increase_1988_2024', ascending=False, inplace=True)

#Display the percentage increase
df_percentage_increase
```

```python
##Creating an interactive bar chart showing percentage increase in military spending in different continents

#Creating an interactive bar chart showing percentage increase in military spending in different continents

import plotly.express as px #Import the plotly library

# Relevant data of percentage increase
data = {
    "Region": [
        "Africa", "Asia", "Europe",
        "North America", "South America"
    ],
    "Percent_Increase": [147.3, 374.6, 3.1, 21.9, 30.0]
}

df = pd.DataFrame(data)


# Plotting
fig = px.bar(
    df,
    x="Region",
    y="Percent_Increase",
    color="Region",
    hover_data=["Percent_Increase"],
    title="Percentage Increase in Military Expenditures by Region (1988â€“2024)",

)

fig.show()
```

```python
##Creating an interactive geo-spatial map of the percentage increase in military spending

# Create the dataframe with the provided data
data = {
    "Region": [
        "Africa", "Asia", "Europe",
        "North America", "South America"
    ],
    "Percent_Increase": [147.3, 374.6, 3.1, 21.9, 30.0]
}

df = pd.DataFrame(data)

# Create a mapping between regions and ISO codes for the choropleth map
# This maps the regions to their corresponding ISO alpha-3 codes
iso_region_mapping = {
    'Africa': ['DZA', 'AGO', 'BEN', 'BWA', 'BFA', 'BDI', 'CMR', 'CPV', 'CAF', 'TCD',
               'COM', 'COG', 'COD', 'DJI', 'EGY', 'GNQ', 'ERI', 'ETH', 'GAB', 'GMB',
               'GHA', 'GIN', 'GNB', 'CIV', 'KEN', 'LSO', 'LBR', 'LBY', 'MDG', 'MWI',
               'MLI', 'MRT', 'MUS', 'MAR', 'MOZ', 'NAM', 'NER', 'NGA', 'RWA', 'STP',
               'SEN', 'SYC', 'SLE', 'SOM', 'ZAF', 'SSD', 'SDN', 'SWZ', 'TZA', 'TGO',
               'TUN', 'UGA', 'ZMB', 'ZWE'],
    'Asia': ['AFG', 'ARM', 'AZE', 'BHR', 'BGD', 'BTN', 'BRN', 'KHM', 'CHN', 'CYP',
             'GEO', 'IND', 'IDN', 'IRN', 'IRQ', 'ISR', 'JPN', 'JOR', 'KAZ', 'KWT',
             'KGZ', 'LAO', 'LBN', 'MYS', 'MDV', 'MNG', 'MMR', 'NPL', 'PRK', 'OMN',
             'PAK', 'PSE', 'PHL', 'QAT', 'SAU', 'SGP', 'KOR', 'LKA', 'SYR', 'TWN',
             'TJK', 'THA', 'TUR', 'TKM', 'ARE', 'UZB', 'VNM', 'YEM'],
    'Europe': ['ALB', 'AND', 'AUT', 'BLR', 'BEL', 'BIH', 'BGR', 'HRV', 'CZE', 'DNK',
               'EST', 'FIN', 'FRA', 'DEU', 'GRC', 'HUN', 'ISL', 'IRL', 'ITA', 'LVA',
               'LIE', 'LTU', 'LUX', 'MKD', 'MLT', 'MDA', 'MCO', 'MNE', 'NLD', 'NOR',
               'POL', 'PRT', 'ROU', 'RUS', 'SMR', 'SRB', 'SVK', 'SVN', 'ESP', 'SWE',
               'CHE', 'UKR', 'GBR', 'VAT'],
    'North America': ['ATG', 'BHS', 'BRB', 'BLZ', 'CAN', 'CRI', 'CUB', 'DMA', 'DOM',
                      'SLV', 'GRD', 'GTM', 'HTI', 'HND', 'JAM', 'MEX', 'NIC', 'PAN',
                      'KNA', 'LCA', 'VCT', 'TTO', 'USA'],
    'South America': ['ARG', 'BOL', 'BRA', 'CHL', 'COL', 'ECU', 'GUY', 'PRY', 'PER',
                      'SUR', 'URY', 'VEN']
}

# Create a new dataframe that maps each country to its region's percent increase
country_data = []
for region, countries in iso_region_mapping.items():
    percent_increase = df[df['Region'] == region]['Percent_Increase'].values[0]
    for country in countries:
        country_data.append({
            'ISO_Code': country,
            'Region': region,
            'Percent_Increase': percent_increase
        })

country_df = pd.DataFrame(country_data)

# Create an interactive choropleth map
fig = px.choropleth(
    country_df,
    locations='ISO_Code',
    color='Percent_Increase',
    hover_name='Region',
    hover_data=['Percent_Increase'],
    color_continuous_scale='PiYG',  # You can change this to other scales like 'Plasma', 'Inferno', etc.
    projection='natural earth',
    title='Regional Percentage Increases in Military Spending Across the World',
    labels={'Percent_Increase': 'Percentage Increase (%)'}
)

# Customize the layout
fig.update_layout(
    coloraxis_colorbar=dict(
        title='Percentage Increase (%)'
    ),
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='equirectangular'
    )
)

# Show the map
fig.show()
```
